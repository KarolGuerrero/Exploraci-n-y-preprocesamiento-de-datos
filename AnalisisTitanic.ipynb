```python
import pandas as pd

df_titanic = pd.read_csv('titanic.csv')
display(df_titanic.head())
```

## ¿Cuál es la tasa de supervivencia general de los pasajeros del Titanic?

```python
tasa_supervivencia = df_titanic['Survived'].value_counts(normalize=True).rename_axis('Sobrevivió').reset_index(name='Proporción')
tasa_supervivencia['Sobrevivió'] = tasa_supervivencia['Sobrevivió'].map({0: 'No', 1: 'Sí'})
tasa_supervivencia.style.format({'Proporción': '{:.2%}'}).set_caption('Tasa de Supervivencia General')
```

## ¿Cómo varía la tasa de supervivencia según el sexo de los pasajeros?

```python
ct_sexo = pd.crosstab(df_titanic['Sex'], df_titanic['Survived'], margins=True)
display(ct_sexo)

tasa_sexo = df_titanic.groupby('Sex')['Survived'].mean().reset_index()
tasa_sexo['Survived'] = tasa_sexo['Survived'].map('{:.2%}'.format)
tasa_sexo = tasa_sexo.rename(columns={'Sex': 'Sexo', 'Survived': 'Tasa de Supervivencia'})
tasa_sexo.style.set_caption('Tasa de Supervivencia por Sexo')
```

## Existe una relación entre la clase del boleto y la probabilidad de sobrevivir?

```python
ct_clase = pd.crosstab(df_titanic['Pclass'], df_titanic['Survived'], margins=True)
display(ct_clase)

tasa_clase = df_titanic.groupby('Pclass')['Survived'].mean().reset_index()
tasa_clase['Survived'] = tasa_clase['Survived'].map('{:.2%}'.format)
tasa_clase = tasa_clase.rename(columns={'Pclass': 'Clase', 'Survived': 'Tasa de Supervivencia'})
tasa_clase.style.set_caption('Tasa de Supervivencia por Clase')
```

## ¿Cuál es la probabilidad de sobrevivir si es niño o adulto?

```python
def categorize_age(age):
    if pd.isna(age):
        return 'Unknown' 
    elif age < 18:
        return 'Child'
    else:
        return 'Adult'
    
df_titanic['AgeGroup'] = df_titanic['Age'].apply(categorize_age)
    
ct_age = pd.crosstab(df_titanic['AgeGroup'], df_titanic['Survived'], margins=True)
display(ct_age)

tasa_edad = df_titanic.groupby('AgeGroup')['Survived'].mean().reset_index()
tasa_edad['Survived'] = tasa_edad['Survived'].map('{:.2%}'.format)
tasa_edad = tasa_edad.rename(columns={'AgeGroup': 'Grupo de edad', 'Survived': 'Tasa de Supervivencia'})
tasa_edad.style.set_caption('Tasa de Supervivencia por Clase')
```
